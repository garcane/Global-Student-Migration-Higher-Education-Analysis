```python


# Filter for placed students with positive salary
df_placed = df[(df['placement_status'] == 'Placed') & (df['starting_salary_usd'] > 0)]

# Select relevant columns
gpa_salary = df_placed[['gpa_or_score', 'starting_salary_usd']]

# Correlation
correlation = gpa_salary['gpa_or_score'].corr(gpa_salary['starting_salary_usd'])
print(f"\nCorrelation between GPA/Score and Starting Salary: {correlation:.4f}")

```

```python

# Bin GPA into ranges (e.g., 2.5-3.0, 3.0-3.5, 3.5-4.0)
# Define finer bins and labels for GPA ranges (e.g., 0.25 increments)
bins = [2.0, 2.25, 2.5, 2.75, 3.0, 3.25, 3.5, 3.75, 4.0]
labels = ['2.00-2.25', '2.25-2.50', '2.50-2.75', '2.75-3.00', '3.00-3.25', '3.25-3.50', '3.50-3.75', '3.75-4.00']
gpa_salary['gpa_bin'] = pd.cut(gpa_salary['gpa_or_score'], bins=bins, labels=labels, include_lowest=True)

# Calculate average salary per GPA bin
avg_salary_by_gpa = gpa_salary.groupby('gpa_bin')['starting_salary_usd'].mean().reset_index()

# Bar chart
plt.figure(figsize=(10, 6))
plt.bar(avg_salary_by_gpa['gpa_bin'], avg_salary_by_gpa['starting_salary_usd'], color='skyblue', edgecolor='black')
plt.xlabel('GPA Range')
plt.ylabel('Average Starting Salary (USD)')
plt.title('Average Starting Salary by GPA Range (Placed Students Only)')
plt.grid(True, axis='y', alpha=0.3)

# Add value labels on top of bars
for i, v in enumerate(avg_salary_by_gpa['starting_salary_usd']):
    plt.text(i, v + 1000, f'${v:,.0f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()  # Or plt.savefig('gpa_vs_salary_bar.png') to save as image
```
